```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating 100 points in third quadrant of a square (0,0)(0,4)
cluster1 = np.random.uniform(low = 0,high = 2,size = (100,2))
#creating 100 points in first quadrant
cluster2 = np.random.uniform(low = 2,high = 4,size = (100,2))
```

```python
plt.scatter([i[0] for i in cluster1],[i[1] for i in cluster1])
plt.scatter([i[0] for i in cluster2],[i[1] for i in cluster2])
plt.show()
```

```python
#mean of cluster1 points
mean1 = np.mean(cluster1,axis = 0)
#mean of cluster2 points
mean2 = np.mean(cluster2,axis = 0)
print(mean1)
```

```python
plt.scatter([i[0] for i in cluster1],[i[1] for i in cluster1])
plt.scatter([i[0] for i in cluster2],[i[1] for i in cluster2])
plt.scatter(mean1[0],mean1[1],color = 'r')
plt.scatter(mean2[0],mean2[1],color = 'b')
plt.show()
```

```python
#calculating distance from a test point
def testing(test_points):
    dist1 = np.sqrt(np.sum(np.square(np.subtract(mean1,test_points))))
    dist2 = np.sqrt(np.sum(np.square(np.subtract(mean2,test_points))))
    if dist1<=dist2:
        print(test_points,"belong to the cluster1")
    else:
        print(test_points,"belong to the cluster2")
```

```python
#testing
testing(np.random.uniform(0,4,(2)))
```
