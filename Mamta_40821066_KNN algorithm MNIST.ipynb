```python
from sklearn.datasets import fetch_openml
import numpy as np
mnist = fetch_openml('mnist_784', version=1)
mnist['frame']
```

```python
mnist 
```

```python
test, y_test = mnist["data"][60000:],mnist["target"][60000:]
test
```

```python
training_data = mnist['frame'][:60000].to_numpy()
testing_data = test.to_numpy()
test_labels = y_test.to_numpy()
training_data
```

```python
labels = np.unique(training_data[:,-1])
print(labels)
```

```python
#creating clusters
clusters = {}
for num in labels:
     clusters[num]= training_data[np.where(training_data[:,-1]==num)][:,0:-1]
print(clusters)
```

```python
#mean of clusters
a = [f"mean{x}" for x in labels]
b = [np.mean(clusters[digit],axis = 0) for digit in labels]
means = dict(zip(a,b))
means
```

```python
#calculating distance from a test point
def testing(test_points):
    x = [digit for digit in clusters.keys()]    #making the keys of the distance dictionary same as keys of cluster dictionary
    y = []
    for digit in means.keys():
        y.append(np.sqrt(np.sum(np.square(np.subtract(means[digit],test_points)))))  #distance from each cluster
    distance = dict(zip(x,y))
    for key in distance.keys():
        if distance[key]==min(distance.values()):
            return key
```

```python
#listing the output of our KNN algorithm 
mlp_output = np.empty(0)
for data_point in testing_data:
    mlp_output = np.append(mlp_output,testing(data_point))
print(mlp_output)
```

```python
#calculating the accuracy
correct = 0
incorrect = 0
for i in range(len(mlp_output)):
    if mlp_output[i]==test_labels[i]:
        correct+=1
    else:
        incorrect+=1
print('correct results = ',correct)
print('incorrect results = ',incorrect)
accuracy = np.multiply(np.divide(correct,len(test_labels)),100)
print(f"Accuracy of our KNN algorithm for MNIST data is = {accuracy} %")
```

```python
#calculating class wise accuracy
for i in np.unique(test_labels):
    correct,incorrect=0,0
    for j in range(len(mlp_output)):
        if test_labels[j]==i:
            if mlp_output[j]==test_labels[j]:
                correct+=1
            else:
                incorrect+=1
    print(f'Total of number of {i} class = ',np.count_nonzero(test_labels==i))
    print(f'correct results for {i} class = ',correct)
    print(f'incorrect results for {i} class = ',incorrect)
    accuracy = np.multiply(np.divide(correct,np.count_nonzero(test_labels==i)),100)
    print(f"Accuracy of our KNN algorithm for MNIST data with {i} class is = {accuracy} %")
```
